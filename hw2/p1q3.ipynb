```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = []
watts = []
freq = []
max_temp = []

with open('log_q3.txt') as f:
    lines = f.readlines()
    time = [line.split()[0] for line in lines]
    watts = [line.split()[1] for line in lines]
    freq = [line.split()[2] for line in lines]
    max_temp = [line.split()[3] for line in lines]
    
```

```python
time = time[1:]
watts = watts[1:]
freq = freq[1:]
max_temp = max_temp[1:]
```

```python
for i in range(len(watts)):
    time[i] = float(time[i])
    watts[i] = float(watts[i])
    freq[i] = float(freq[i]) / 1000000
    max_temp[i] = float(max_temp[i])
```

```python
print(max_temp)
```

```python
plt.plot(watts)
plt.title('System Power Over Time: Blacksholes')
plt.xlabel('Time (s)')
plt.ylabel('Power (W)')
plt.show()
```

```python
plt.plot(freq)
plt.title('Frequency of Big Cores Over Time: Blacksholes')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (GHz)')
plt.show()
```

```python
plt.plot(max_temp)
plt.title('System Max Big Temp Over Time: Blacksholes')
plt.xlabel('Time (s)')
plt.ylabel('Temp. (degrees celcius)')
plt.show()
```
