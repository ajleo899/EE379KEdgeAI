# Blacksholes

```python
time = []
watts = []
freq = []
max_temp = []

with open('log_q3.txt') as f:
    lines = f.readlines()
    time = [line.split()[0] for line in lines]
    watts = [line.split()[1] for line in lines]
    freq = [line.split()[2] for line in lines]
    max_temp = [line.split()[3] for line in lines]
    
```

```python
time = time[41:961]
watts = watts[41:961]
freq = freq[41:961]
max_temp = max_temp[41:961]
```

```python
for i in range(len(watts)):
    time[i] = float(time[i])
    watts[i] = float(watts[i])
    freq[i] = float(freq[i]) / 1000000
    max_temp[i] = float(max_temp[i])
```

```python
print(max_temp)
```

```python
plt.plot(watts)
plt.title('System Power Over Time: Blacksholes')
plt.xlabel('Time (s)')
plt.ylabel('Power (W)')
plt.show()
```

```python
plt.plot(freq)
plt.title('Frequency of Big Cores Over Time: Blacksholes')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (GHz)')
plt.show()
```

```python
plt.plot(max_temp)
plt.title('System Max Big Temp Over Time: Blacksholes')
plt.xlabel('Time (s)')
plt.ylabel('Temp. (degrees celcius)')
plt.show()
```

# Bodytype

```python
time = []
watts = []
freq = []
max_temp = []

with open('log_q3_2.txt') as f:
    lines = f.readlines()
    time = [line.split()[0] for line in lines]
    watts = [line.split()[1] for line in lines]
    freq = [line.split()[2] for line in lines]
    max_temp = [line.split()[3] for line in lines]
    

time = time[1:]
watts = watts[1:]
freq = freq[1:]
max_temp = max_temp[1:]


for i in range(len(watts)):
    time[i] = float(time[i])
    watts[i] = float(watts[i])
    freq[i] = float(freq[i]) / 1000000
    max_temp[i] = float(max_temp[i])

print(max_temp)

plt.plot(watts)
plt.title('System Power Over Time: Bodytrack')
plt.xlabel('Time (s)')
plt.ylabel('Power (W)')
plt.show()

plt.plot(freq)
plt.title('Frequency of Big Cores Over Time: Bodytrack')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (GHz)')
plt.show()

plt.plot(max_temp)
plt.title('System Max Big Temp Over Time: Bodytrack')
plt.xlabel('Time (s)')
plt.ylabel('Temp. (degrees celcius)')
plt.show()
```

# Bodytrack calculations

```python
# average power 
average_power = sum(watts) / len(watts)
print(average)
```

```python
# average max temp
average_max_temp = sum(max_temp) / len(max_temp)
print(average_max_temp)
```

```python
# max temp
highest_temp = 0

for temp in max_temp:
    if temp > highest_temp:
        highest_temp = temp
        
print(highest_temp)
```
